## Square Root Transform

```python
from matplotlib import pyplot
```

```python
s2=[k**2 for k in range(1,100)]
pyplot.figure(1)
pyplot.subplot(211)
pyplot.plot(s2);

pyplot.subplot(212)
pyplot.hist(s2);
```

```python
from numpy import sqrt
transforms2=sqrt(s2)
```

```python

pyplot.subplot(211)
pyplot.plot(transforms2);
pyplot.subplot(212)
pyplot.hist(transforms2);
```

## Log Transform

```python
from math import exp

s3=[exp(k) for k in range(1,100)]
pyplot.figure(1)

pyplot.subplot(211)
pyplot.plot(s3);


pyplot.subplot(212)
pyplot.hist(s3);
```

```python
from numpy import log

transform1 = log(s3)
```

```python
import numpy as np

transform2 = np.log(s3)
```

we can use  from numpy import log and then use the function log() or alternative

we can use  import numpy as np and then call the function log as np.log()

```python
[transform1, transform2]
```

```python
pyplot.figure(1)

pyplot.subplot(211)
pyplot.plot(transform1);


pyplot.subplot(212)
pyplot.hist(transform1);
```

## Box-Cox Transform

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import boxcox
```

```python
s5= read_csv('../data/airline-passengers.csv', header=0, index_col=0, parse_dates=True,squeeze=True)
```

```python
s5.head()
```

```python
df5=pd.DataFrame(s5)
```

```python
df5.head()
```

```python
df5['passangers_bc']=boxcox(df5['Passengers'], lmbda=0)
df5.head()
```

```python
plt.subplot(211)
df5['passangers_bc'].plot()
plt.subplot(212)
df5['passangers_bc'].hist()
```

```python
df6=pd.DataFrame(s5)
df6['passangers_bc'], lam=boxcox(df6['Passengers'])
df6.head()
lam
```

```python
plt.subplot(211)
df6['passangers_bc'].plot()
plt.subplot(212)
df6['passangers_bc'].hist()
```
