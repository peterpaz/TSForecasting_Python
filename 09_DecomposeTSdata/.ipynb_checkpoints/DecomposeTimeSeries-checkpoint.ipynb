# Additive Decomposition

```python
from random import randrange
from statsmodels.tsa.seasonal import seasonal_decompose

series=[i+randrange(10) for i in range(1,100)]
result=seasonal_decompose(series, model='additive', period=1)
result.plot();
```

# Multiplicative Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose

series=[i**2 for i in range(1,100)]
result=seasonal_decompose(series, model='multiplicative', period=1)
result.plot();
```

- It is important to treat decomposition as a potentially useful analysis tool, but to consider exploring the many different ways it could be applied for your problem, such as on data after it has been transformed or on residual model errors
