```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
s1 = pd.read_csv('../data/daily-total-female-births.csv', header=0, index_col=0, parse_dates=True,
    squeeze=True)
```

```python
s1
```

```python
rolling=s1.rolling(window=3) # here is only to define the windown
```

```python
rolling_mean=rolling.mean()
rolling_mean
```

```python
s1.plot()
```

```python
rolling_mean.plot(color='red')
```

```python
s1.plot()
rolling_mean.plot(color='red')
```

```python
s1[:100].plot()
rolling_mean[:100].plot(color='red')
```
